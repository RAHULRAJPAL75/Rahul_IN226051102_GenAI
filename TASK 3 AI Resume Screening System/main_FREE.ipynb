## Step 6: Initialize LLM and Build Pipeline

````python
llm = build_llm(
    api_key=os.environ["GOOGLE_API_KEY"],
    provider="gemini",
    model="gemini-pro",
    temperature=0.0
)

screening_pipeline = build_screening_pipeline(llm)

print("✅ Google Gemini LLM initialized (FREE!)")
print("✅ Screening pipeline built!")
````

## Step 7: Run Pipeline for All 3 Candidates

````python
results = []

candidates = [
    {"name": "Priya Sharma (STRONG)", "resume": resume_strong},
    {"name": "Rahul Verma (AVERAGE)", "resume": resume_average},
    {"name": "Amit Kumar (WEAK)", "resume": resume_weak}
]

print("\n" + "="*70)
print("🚀 RUNNING AI RESUME SCREENING PIPELINE")
print("="*70 + "\n")

for candidate in candidates:
    print(f"\n📄 Processing: {candidate['name']}")
    print("-" * 70)
    
    result = screening_pipeline.invoke({
        "resume": candidate["resume"],
        "job_description": job_description,
        "candidate_name": candidate["name"]
    })
    
    results.append(result)
    print(f"✅ Completed: {candidate['name']}\n")

print("\n" + "="*70)
print("✅ ALL CANDIDATES PROCESSED SUCCESSFULLY!")
print("="*70)
print("\n🔍 Check LangSmith dashboard: https://smith.langchain.com/")
````

## Step 8: Display Results

````python
def display_result(result):
    print("\n" + "="*70)
    print(f"📋 CANDIDATE: {result['candidate_name']}")
    print("="*70)
    
    print("\n🔍 STEP 1: EXTRACTED PROFILE")
    print("-" * 70)
    print(result['extracted_profile'])
    
    print("\n🔗 STEP 2: MATCH ANALYSIS")
    print("-" * 70)
    print(result['match_analysis'])
    
    print("\n📊 STEP 3: SCORE RESULT")
    print("-" * 70)
    print(result['score_result'])
    
    print("\n💡 STEP 4: EXPLANATION & RECOMMENDATION")
    print("-" * 70)
    print(result['explanation'])
    print("\n" + "="*70 + "\n")

for result in results:
    display_result(result)
````

## Step 9: Summary Comparison

````python
print("\n" + "="*70)
print("📊 SUMMARY COMPARISON TABLE")
print("="*70 + "\n")

print(f"{'Candidate':<30} {'Total Score':<15} {'Grade':<15}")
print("-" * 70)

for result in results:
    try:
        score_data = json.loads(result['score_result'])
        total_score = score_data.get('total_score', 'N/A')
        grade = score_data.get('grade', 'N/A')
    except:
        total_score = "Check output"
        grade = "Check output"
    
    print(f"{result['candidate_name']:<30} {str(total_score):<15} {str(grade):<15}")

print("\n" + "="*70)
````

````python
print("\n" + "="*70)
print("🔍 LANGSMITH TRACING VERIFICATION")
print("="*70 + "\n")

print("✅ LangSmith tracing is ENABLED")
print(f"✅ Project Name: {os.environ.get('LANGCHAIN_PROJECT', 'Not Set')}")
print(f"✅ Total Runs: {len(results)} candidates processed")
print("\n📌 View traces at: https://smith.langchain.com/")
print("\n📸 IMPORTANT: Take screenshots for submission!")
print("="*70)
````
