## Step 2: Enter API Keys

```python
import os
from getpass import getpass

# Google Gemini API Key
os.environ["GOOGLE_API_KEY"] = getpass("Gemini API Key: ")

# LangSmith API Key
os.environ["LANGCHAIN_TRACING_V2"] = "true"
os.environ["LANGCHAIN_PROJECT"] = "AI-Resume-Test"
os.environ["LANGCHAIN_API_KEY"] = getpass("LangSmith API Key: ")

print("✅ API Keys set!")
```

## Step 3: Quick Test

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

# Initialize Gemini
llm = ChatGoogleGenerativeAI(
    model="gemini-pro",
    temperature=0.0,
    google_api_key=os.environ["GOOGLE_API_KEY"]
)

print("✅ LLM initialized!")

# Simple test
resume = """
Name: Priya Sharma
Experience: 5 years as Data Scientist
Skills: Python, Machine Learning, TensorFlow, AWS
Education: Master's in Data Science
"""

job = """
Required: 3+ years experience, Python, ML, Cloud
"""

prompt = PromptTemplate(
    input_variables=["resume", "job"],
    template="""
Analyze this resume against the job requirements.

Resume: {resume}

Job: {job}

Provide:
1. Score (0-100)
2. Matched skills
3. Recommendation
"""
)

chain = prompt | llm | StrOutputParser()

print("\n🚀 Processing...\n")

result = chain.invoke({"resume": resume, "job": job})

print("="*70)
print("📊 RESULT:")
print("="*70)
print(result)
print("="*70)
print("\n✅ DONE! Check LangSmith: https://smith.langchain.com/")
print("\n💰 Cost: $0.00 (FREE!)")
```
