```python
from langchain_core.prompts import ChatPromptTemplate

def get_chat_prompt(role, topic):
    if role == "teacher":
        system_message = "You are a helpful teacher."
    elif role == "interviewer":
        system_message = "You are a strict interviewer."
    elif role == "motivator":
        system_message = "You are a motivational speaker."
    else:
        system_message = "You are an assistant."

    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_message),
        ("user", "Explain {topic}")
    ])

    return chat_template.format_messages(topic=topic)

messages = get_chat_prompt("teacher", "Neural Networks")
print(messages)
```

```python
def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("Invalid audience! Setting default = beginner")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone! Setting default = formal")
        tone = "formal"

    return audience, tone

aud, tone = validate_inputs("kids", "funny")
print(aud, tone)
```

```python
def generate_prompt(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone)

    if style == "teaching":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} step by step for {audience} in a {tone} tone"
        )

    elif style == "interview":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Generate interview questions on {topic} for {audience} in a {tone} tone"
        )

    elif style == "storytelling":
        template = PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} as a story for {audience} in a {tone} tone"
        )

    else:
        return "Invalid style!"

    return template.format(topic=topic, audience=audience, tone=tone)

print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

```python
from langchain_core.prompts import PromptTemplate

def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        print("Invalid audience! Default set to 'beginner'")
        audience = "beginner"

    if tone not in valid_tone:
        print("Invalid tone! Default set to 'formal'")
        tone = "formal"

    return audience, tone

def get_template(style):

    if style == "teaching":
        return PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} step by step for {audience} in a {tone} tone"
        )

    elif style == "interview":
        return PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Generate 3 interview questions on {topic} for {audience} in a {tone} tone"
        )

    elif style == "storytelling":
        return PromptTemplate(
            input_variables=["topic", "audience", "tone"],
            template="Explain {topic} as a story for {audience} in a {tone} tone"
        )

    else:
        return None

def generate_prompt(topic, audience, tone, style):

    audience, tone = validate_inputs(audience, tone)
    template = get_template(style)

    if template is None:
        return "Invalid style selected!"


    final_prompt = template.format(
        topic=topic,
        audience=audience,
        tone=tone
    )

    return final_prompt
topic = "Neural Networks"
audience = "beginner"
tone = "fun"
style = "storytelling"

output = generate_prompt(topic, audience, tone, style)

print("FINAL GENERATED PROMPT:\n")
print(output)
```
